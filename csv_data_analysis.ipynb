# Loading and cleading the data

```python
import pandas as pd

df = pd.read_csv("100 Sales Records.csv")
print(df.head())  # Preview the first few rows
print(df.info())
print(df.describe())
print(df.isnull().sum())  # Check for missing values
df.dropna(inplace = True) # Eliminate Null 
# Total sales by product
sales_by_product = df.groupby("Item Type")["Total Revenue"].sum().sort_values(ascending=False)
print(sales_by_product)

total_revenue = df.groupby("Region")["Total Revenue"].sum()
total_profit = df.groupby("Country")["Total Profit"].sum().sort_values(ascending=False).head(10)
most_sold_item = df.groupby("Item Type")["Units Sold"].sum().sort_values(ascending=False)
print(f"TOTAL REVENUE: {total_revenue}")
print(f"TOTAL PROFIT: {total_profit}")
print(f"MOST SOLD ITEM: {most_sold_item}")
```
